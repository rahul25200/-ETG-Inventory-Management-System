#  Reading inventory from record.json file

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
records
```

# Purchasing products

```python
print("--------------- Welcome to the shop ! ---------------")

form_run = True

while (form_run):
    ui_prod = input("\nEnter the Product ID you want to purchase: ")

    if ui_prod in records:
        print("-------------------------------------------")
        print("Product: ", records[ui_prod]['name'])
        print("Brand: ", records[ui_prod]['brand'])
        print("Price: ", records[ui_prod]['price'])
        print("Quantity available: ", records[ui_prod]['quant'])
        print("Expiry: ", records[ui_prod]['expiry'])
        print("-------------------------------------------")

        ui_quant = int(input("Enter the quantity: "))

        if (records[ui_prod]['quant'] >= ui_quant):
 
            records[ui_prod]['quant'] = records[ui_prod]['quant'] - ui_quant
    
            print("\n**************** BILLING DETAILS ****************\n")
            print("Total Amount: Rs. " + str(records[ui_prod]['price'] * ui_quant)) 
            print("Thank You. Visit Again\n")
            print("***************************************************")
        
            form_run = False

        else:
            print("\nSorry! We have only " + str(records[ui_prod]['quant']) + " quantity available.")
        
    else:
        print("\nSorry! No product found for ID: ", ui_prod)
        print("Enter valid product ID!")
    
#Updating inventory after purchase

js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

# Updating sales.json file

```python

import time

fd = open("sales.json", 'r')
r = fd.read()
fd.close()

sales = json.loads(r)
num = len(sales) + 1

sales[num] = {'prod_id': ui_prod, 'quant': ui_quant, 'amount': records[ui_prod]['price'] * ui_quant, 'date_time': time.ctime()}
sale = json.dumps(sales)

fd = open("sales.json",'w')
fd.write(sale)
fd.close()
```
